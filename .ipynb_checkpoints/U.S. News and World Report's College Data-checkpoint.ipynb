### U.S. News and World Report's College Data

#### Background

The USNews college data set is retrieved from the website of the text, "Introduction to Statistical Learning" by G. James, D. Witten,  T. Hastie and R. Tibshirani. The dataset includes statistics for a large number of US Colleges from the 1995 issue of the US News and World Report. Statistics on each college (n=777) include number of applications, acceptances, enrollment, students' high school grades, tuition, room and board, proportion of faculty with PhDs, and so on. The goal here is to fit a linear regression model to predict the graduate rate of a school given its attributes. 

### Inspecting data and some EDA

```python
#import libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
%matplotlib inline

#import the college data from the ISLR webpage. 
file_link = 'http://www-bcf.usc.edu/~gareth/ISL/College.csv'
df = pd.read_csv(file_link)
df.head()
```

We will drop the name the college since it is not a helpful feature.

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df.info()
```

Let's convert the private school label into a dummy variable

```python
df = pd.get_dummies(df, drop_first=True)
df.head()
```

```python
df.describe()
```

About 72% of the colleges here were private. The mean graduation rate was 65%; however, the max showed 118%. Let's see if there were any school whose graduate rate was more than 100%, which doesn't make sense.

```python
df[df['Grad.Rate'] > 100]
```

In fact there was one school whose graduate rate was recorded as 118%; we will drop this record since it was most likely a data entry error.

```python
#that school's index is 95
df.drop(95, inplace=True)
```

```python
#let's plot a histogram of graduate rate, which will be our target variable
_ = df['Grad.Rate'].hist()
```

Let's visualize the pairwise relation of these numeric variables with a seaborn PairGrid and plot histograms of each variable along the diagonal.

```python
g = sns.PairGrid(df)
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
```

We can also use a heat map to see the correlations. 

```python
_ = sns.heatmap(df.corr())
```
