Let's visualize each pair of numeric variables with seaborn pair grid. We can add histograms along the diagonal to show the univariate distribution of each numeric variable. Let's also specify income label as the hue argument. 

```python
g = sns.PairGrid(df, hue='income')
_ = g.map_diag(plt.hist)
_ = g.map_offdiag(plt.scatter)
_ = g.add_legend()
```

There were more people making less than 50K than more. It also appears that people with college-level education or higher were much more likely to earn more than 50K. People less 25 years older were likely to make less than 50K. 


Let's see the visualize income label by gender with a few numerical features.

```python
g = sns.FacetGrid(df, col="sex", hue="income")
_ = g.map(plt.scatter, "education-num", "hours-per-week",alpha=.5)
_ = g.add_legend()
```
